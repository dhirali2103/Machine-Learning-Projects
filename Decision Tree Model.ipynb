**Desicion Tree:**

--> Decision Tree is a tree shaped diagream used to determine a course of action. Each branch of the tree represents a possible decision, occurence or reaction.

**Problems that Decision Tree can solve:**

--> Classification: A classification tree will determine a set of logical if-then conditions to classify problems.

Ex: Discriminating between three types of flowers based on certain features.


--> Regression:  A regression tree is used when the target variable is numerical or continous in nature. We fir a regression model to the target variables. Each split is made based on the sum of squared error.

**Important Terms:**

**Entropy ->** Entropy is the measure of randomness or unpredictability in the dataset.

Ex: The Dataset has a very high entropy.

**Information Gain ->** It is the measure of decrease in entropy after the dataset is split.

**Leaf Node ->**Leaf node carries the classification or the decision.

**Root Node ->** The top most decision node is known as the root node.

**Use case:** Loan Repayment Prediction

**Problem statement:** To predict if a customer will repay loan ammount or not using decision tree algorithm in python

```python
# import the necessary packages

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
# Load the data from .csv file.

data = pd.read_csv('Dataset.csv',sep=',',header=0)
```

```python
print("Dataset Length: ",len(data))
print("Dataset Shape: ",data.shape)
```

```python
print("Dataset: ")
data.head()
```

```python
# Rename the specific colums

data = data.rename(columns={"1":"Inital Payment","2":"Last payment","3":"Credit Score","4":"House Number","Unnamed: 5":"Result"})

data.to_csv('Dataset.csv',index=False)
```

```python
print("Dataset: ")
data.head()
```

```python
# Seperating the Target variable

X = data.iloc[:,0:4]
y = data.iloc[:,5]
```

```python
# Spliting Dataset into test and train

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# Function to perform training with entropy
clf_entropy = DecisionTreeClassifier(criterion="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
clf_entropy.fit(X_train,y_train)
```

```python
# Function to make predictions

y_pred_en = clf_entropy.predict(X_test)
print(y_pred_en)
```

```python
# Checking the Accuracy

print("Accuracy is",accuracy_score(y_test,y_pred_en)*100)
```

-> The bank can use this model to decide whether it should approve loan request from a particular customer or not.
